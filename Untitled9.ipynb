```python
import os, numpy as np, matplotlib.pyplot as plt

# 1) Load artifacts (or train dummy if missing)
if not os.path.exists("training_data.npz"):
    print("training_data.npz not found — generating one via training script…")
    !python penalty_shootout_ql.py

data = np.load("training_data.npz", allow_pickle=True)
Q = data["Q"]; returns = data["returns"]; moving_avg = data["moving_avg"]; epsilons = data["epsilons"]; ACTIONS = list(data["actions"])
print("Loaded training_data.npz:", Q.shape, "states×actions")

# 2) Learning curves
plt.figure(); plt.plot(returns, label="Episode goals"); plt.plot(moving_avg, label="MA(50)"); plt.xlabel("Episode"); plt.ylabel("Goals"); plt.title("Learning Curve"); plt.legend(); plt.tight_layout(); plt.show()
plt.figure(); plt.plot(epsilons); plt.xlabel("Episode"); plt.ylabel("ε"); plt.title("Exploration Schedule"); plt.tight_layout(); plt.show()

# 3) Greedy action map across state space
KICKS=5; MAX_DIFF=KICKS; COLS=2*MAX_DIFF+1; ROWS=KICKS+1
def idx(k,d): return (k*COLS)+(d+MAX_DIFF)
greedy = np.argmax(Q, axis=1)
grid = np.zeros((ROWS, COLS), dtype=int)
for k in range(ROWS):
    for d in range(-MAX_DIFF, MAX_DIFF+1):
        grid[k, d+MAX_DIFF] = greedy[idx(k,d)]
plt.figure(); plt.imshow(grid, aspect="auto"); plt.xticks(range(COLS), [str(d) for d in range(-MAX_DIFF,MAX_DIFF+1)]); plt.yticks(range(ROWS), [str(k) for k in range(ROWS)])
plt.xlabel("Score diff"); plt.ylabel("Kick index"); plt.title("Greedy Action Map (indices)"); plt.colorbar(); plt.tight_layout(); plt.show()
print("Legend:", {i:a for i,a in enumerate(ACTIONS)})

# 4) Q-values slice (kick=2 across score diffs)
fixed_k=2
M = np.zeros((len(ACTIONS), COLS))
for a in range(len(ACTIONS)):
    for d in range(-MAX_DIFF, MAX_DIFF+1):
        M[a, d+MAX_DIFF] = Q[idx(fixed_k,d), a]
plt.figure(); plt.imshow(M, aspect="auto"); plt.yticks(range(len(ACTIONS)), ACTIONS); plt.xticks(range(COLS), [str(d) for d in range(-MAX_DIFF,MAX_DIFF+1)])
plt.xlabel("Score diff"); plt.ylabel("Action"); plt.title(f"Q-values at Kick {fixed_k}"); plt.colorbar(); plt.tight_layout(); plt.show()
```
